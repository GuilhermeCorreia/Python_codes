```python
document = docx.Document(docpath)
```

```python
today = datetime.datetime.today()
```

```python
months_dictionary = {
    1:'janeiro',
    2:'fevereiro',
    3:'março',
    4:'abril',
    5:'maio',
    6:'junho',
    7:'julho',
    8:'agosto',
    9:'setembro',
    10:'outubro',
    11:'novembro',
    12:'dezembro'
}

substitutions = {
    2:str(today.day),
    4:str(months_dictionary[today.month]),
    6:str(today.year)+'.'
}
```

```python
for paragraf in document.paragraphs:
    if 'São Paulo' in paragraf.text:
        old_text = paragraf.text.split()
        new_text = old_text
        for key in substitutions:
            new_text[key] = substitutions[key]
        new_text = ' '.join(new_text)
        
        paragraf.text = new_text
        print(paragraf.text)
        
    document.save(docpath)
```

```python
wdFormatPDF = 17

word = comtypes.client.CreateObject('word.application')

docpath

doc = word.Documents.Open(docpath)

doc.SaveAs(docpath[:-5]+ '.pdf', FileFormat=wdFormatPDF)

doc.Close()
word.Quit()


pdf_file = docpath[:-5]+ '.pdf'
```
