```python
def conta_frq(coluna):
    conta = df.groupby(by=[coluna]).sum().drop(['preco','qt','qtd_original'], axis=1).reset_index()
    conta['frq'] = conta['count']/conta['count'].sum()*100
    conta = conta.sort_values(by=['frq'], ascending=False)
    conta['frq'] = conta['frq'].map('{:,.2f} %'.format)

    return conta
```
