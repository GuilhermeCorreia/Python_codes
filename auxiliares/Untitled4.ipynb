```python
for v in variaveis:
    subfolder = globals()[v]
    print(subfolder)
    message = subfolder.Items
    for msg in message:
        try:
            recipients_list = []
            subject = msg.Subject
            sender = msg.SenderEmailAddress
            body = msg.Body
            for word in body.split(' '):
                if '@' in word:
                    recipients_list.append(word)
            e_mail_list.append({'subject':subject, 'sender':sender, 'body':recipients_list})
        except:
            continue
```
