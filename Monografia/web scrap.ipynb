```python
import pandas as pd
import numpy as np
import datetime

from bs4 import BeautifulSoup
import urllib2
import requests
```

```python
url = 'http://valor.com.br/'
```

```python
page = urllib2.urlopen(url)
```

```python
soup = BeautifulSoup(page,'html.parser')
```

```python
all_links = soup.find_all('a', attrs={'class':'comments'})
lst_links = []
for link in all_links:
    lst_links.append(url + link.get('href'))
```

```python
lst_links
```

```python
n_url = lst_links[0]
n_page = urllib2.urlopen(n_url)
n_soup = BeautifulSoup(n_page,'html.parser')
```

```python
texto = n_soup.find(attrs={'class':'n-content'}).get_text().split()
```

```python
res = [item.encode("ascii","ignore") for item in texto]
```
