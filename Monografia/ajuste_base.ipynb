```python
def zscore(df):
    df = df.T
    df_p = (df-df.mean())/df.std()
    return df_p.T
```

```python
market = pd.read_excel('regressao_monografia.xlsx')
```

```python
market = market.set_index('data')
```

```python
market_score = zscore(market).fillna(0.0)
```
