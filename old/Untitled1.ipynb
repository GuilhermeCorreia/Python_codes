```python
%%time


for file in os.listdir(pasta):
    
    if 'Cayman - UBS' in file:
        try:
            ubs_cayman
        except:
            ubs_cayman = pd.read_excel(pasta+'\\'+file)
        else:
            ubs_cayman_1 = pd.read_excel(pasta+'\\'+file)
            ubs_cayman = pd.concat([ubs_cayman, ubs_cayman_1], axis=0)
            
    elif 'Cayman - MS' in file:
        try:
            ms_cayman
        except:
            ms_cayman = pd.read_excel(pasta+'\\'+file)
        else:
            ms_cayman_1 = pd.read_excel(pasta+'\\'+file)
            ms_cayman = pd.concat([ms_cayman, ms_cayman_1], axis=0)
            
            
    elif 'INT C' in file:
        try:
            intc
        except:
            intc = pd.read_excel(pasta+'\\'+file)
        else:
            intc_1 = pd.read_excel(pasta+'\\'+file)
            intc = pd.concat([intc, intc_1], axis=0)        
            
    else:
        print (file)
```

```python
ubs_cayman['tipo'] = 'ubs_cayman'
ms_cayman['tipo'] = 'ms_cayman'
intc['tipo'] = 'intc'
```

```python
relat = pd.concat([ubs_cayman, ms_cayman, intc], axis=0)
```

```python
relat['Position Date'] = pd.to_datetime(relat['Position Date'])
```

```python
relat['month'] = relat['Position Date'].dt.month
relat['year'] = relat['Position Date'].dt.year
```

```python
relat = relat[((relat.month==4)|(relat.month==3))&(relat.year==2018)]
```

```python
relat
```

```python
relat['accrual_adjusted'] = np.where(relat['Position Date'].dt.weekday==4, relat['Daily Loan Cost']*3, relat['Daily Loan Cost'])
```

```python
batmento = relat.groupby(by=['tipo', 'month', 'Currency']).sum().drop(['Account Name', 'Daily Loan Cost', 'Shares Quantity',
                                                           'year'], axis=1)
```
