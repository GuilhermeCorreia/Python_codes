```python
df = c.query('select codigo, date, ticker, sequencia_vencimento_futuro, preabe,premin, premax, '\
                    'premed,data_limite_negociacao,'\
                        'preult, cotacao_ajuste, data_liquidacao_financeira from assets_precofuturo where codigo="DI1" or \
                        codigo="DAP" or codigo="IND"')
```

```python
df=df[(df['sequencia_vencimento_futuro']<3)&(df['sequencia_vencimento_futuro']!=0)]
```

```python
df[df['codigo']=='DAP']
```

```python
df_di = df[df['codigo']=='DI1']
```

```python
data_venc = df_di['data_liquidacao_financeira'].drop_duplicates()
```

```python
pvt= pd.pivot_table(df_di, values=['preabe','preult','premin','premax','cotacao_ajuste'],
                             index='date', columns=['sequencia_vencimento_futuro']).dropna()
```

```python
pvt['contrato']=np.where(pvt.index.isin(data_venc),2,1)
pvt['ajuste'] = np.where((pvt['contrato']==2)&(pvt['contrato'].shift(1)==1),
                                      pvt['cotacao_ajuste'][2].shift(1)/pvt['cotacao_ajuste'][1].shift(1),1)
pvt = pvt.sort_index(ascending=False)
pvt['fator'] = pvt['ajuste'].cumprod()
pvt = pvt.sort_index(ascending=True)

pvt['preult_ajustado'] = np.where(pvt['contrato']==1, pvt['cotacao_ajuste'][1]*pvt['fator'],
                                  pvt['cotacao_ajuste'][2]*pvt['fator']/pvt['ajuste'])
pvt['preabe_ajustado'] = np.where(pvt['contrato']==1, pvt['preabe'][1]*pvt['fator'],
                                  pvt['preabe'][2]*pvt['fator']/pvt['ajuste'])
```
