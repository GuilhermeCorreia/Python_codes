# O código abaixo é para atualizar os dados da Anbima

```python
data=pd.to_datetime(datetime.date.today())
l_days=[]
count=0
while count<5:
    data=buscar_dia_anterior(data)
    l_days.append(data)
    count+=1
l_days=pd.to_datetime(l_days)
```

```python
for data in l_days: #alterar a data de referência
    try:
        df_anbima
    except:
        df_anbima = imp.ImportAnbima(data).dados_anbima()
    else:
        dn = imp.ImportAnbima(data).dados_anbima()
        frames = [df_anbima,dn]
        df_anbima=pd.concat(frames)
```

```python
df_anbima
```

```python
df=df_anbima[['data','titulo','codigo_selic','emissao','vencimento','periodo','compra','venda','tx_indicativa','pu',
              'intervalo_indicativo_minimo_d0','intervalo_indicativo_maximo_d0','intervalo_indicativo_minimo_d1',
              'intervalo_indicativo_maximo_d1']]

df[['compra','venda']]=df[['compra','venda']].apply(pd.to_numeric,errors='coerce').fillna(0.0)
df=df.drop_duplicates(subset=['data','titulo','vencimento','periodo'])
df=df.sort_values(by=['data','titulo','vencimento']).reset_index(drop=True)

df['data_ajuste']=df['data'].apply(pd.to_datetime,format='%Y-%m-%d')
df_tb['data_ajuste']=df_tb['data'].astype(str).str.split(expand=True)[0].apply(pd.to_datetime,format='%Y-%m-%d')

df['check']=df['vencimento'].astype(str)+df['titulo'].astype(str)+df['periodo'].astype(str)+df['data_ajuste'].astype(str)
df_tb['check']=df_tb['vencimento'].astype(str)+df_tb['titulo'].astype(str)+df_tb['periodo'].astype(str)+df_tb['data_ajuste'].astype(str)

df = df[df['check'].isin(df_tb['check'])==False].drop(['check','data_ajuste'],1)
df
```

```python
df.to_sql(con=connection, name='map_rf_cmarkets', if_exists='append', index=False)
```
