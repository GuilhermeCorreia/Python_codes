```python
for file in os.listdir(pasta):
    if 'Stock Loan Fee' in file:
        try:
            chile
        except:
            chile = pd.read_excel(pasta+'\\'+file)
        else:
            chile1 = pd.read_excel(pasta+'\\'+file)
            chile = pd.concat([chile,chile1], axis=0)
```

```python
chile = chile.reset_index(drop=True)
```

```python
chile
```

```python
chile_grouped = chile.groupby(by=['CLIENT','BORROW ID','CURRENT DATE','PRODUCT DESCRIPTION']).sum()\
                .reset_index()[['CLIENT','PRODUCT DESCRIPTION','BORROW ID','CURRENT DATE','CURRENT QUANTITY']].rename(columns={
                    'CURRENT QUANTITY':'CURRENT QUANTITY GROUPED'
                }).sort_values(by=['CLIENT','PRODUCT DESCRIPTION','BORROW ID','CURRENT DATE'])
```

```python
chile_grouped
```

```python
chile_grouped['MOV']=np.where((chile_grouped['CLIENT']==chile_grouped['CLIENT'].shift(1))\
                              &(chile_grouped['PRODUCT DESCRIPTION']==chile_grouped['PRODUCT DESCRIPTION'].shift(1))\
                              &(chile_grouped['BORROW ID']==chile_grouped['BORROW ID'].shift(1)),\
                              chile_grouped['CURRENT QUANTITY GROUPED']-chile_grouped['CURRENT QUANTITY GROUPED'].shift(-1),0)
```

```python
pvt_control = pd.pivot_table(chile_grouped, values=['CURRENT QUANTITY GROUPED'],
                            index=['CLIENT','BORROW ID','PRODUCT DESCRIPTION'], columns='CURRENT DATE').fillna(0)
```
